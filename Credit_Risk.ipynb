# Here we can see that missing values are present in AMT_GOODS_PRICE,NAME_TYPE_SUITE and OCCUPATION_TYPE columns so we have to treat for missing values for better understanding the data

```python
df['AMT_GOODS_PRICE'].fillna(df['AMT_GOODS_PRICE'].mean(),inplace=True)
```

```python
df['NAME_TYPE_SUITE'].fillna(df['NAME_TYPE_SUITE'].mode()[0],inplace=True)
```

```python
df['OCCUPATION_TYPE'].fillna('Not Specified',inplace=True)
```

```python
df['CNT_FAM_MEMBERS'].fillna(df['CNT_FAM_MEMBERS'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```
